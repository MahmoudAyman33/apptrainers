```python
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
df = pd.read_csv('traffic.csv')
df
```

```python
df.info()
```

1. Vehicle Distribution Analysis:

```python
fig, axes = plt.subplots(2, 2, figsize=(12, 8))
vehicles = ['CarCount', 'BikeCount', 'BusCount', 'TruckCount']
colors = ['blue', 'green', 'red', 'orange']
for ax, vehicle, color in zip(axes.flatten(), vehicles, colors):
    sns.histplot(df[vehicle], bins=20, kde=True, ax=ax, color=color)
    ax.set_title(f"Distribution of {vehicle}")
plt.tight_layout()
plt.show()
```

2. Traffic Situation Distribution:

```python
plt.figure(figsize=(8, 8))
df['Traffic Situation'].value_counts().plot.pie(autopct='%1.1f%%', colors=['green', 'orange', 'red','blue'])
plt.title("Traffic Situation Distribution")
plt.ylabel("")
plt.show()
```

# 3. Variation by Day of the Week


```python
plt.figure(figsize=(12, 6))
sns.boxplot(data=df, x="Day of the week", y="Total", palette="coolwarm")
plt.title("Vehicle Count Variation by Day of the Week")
plt.xticks(rotation=45)
plt.show()
```

# 4-7. Vehicle Counts vs. Traffic Situation


```python
# 4-7. Vehicle Counts vs. Traffic Situation
fig, axes = plt.subplots(2, 2, figsize=(12, 8))
for ax, vehicle in zip(axes.flatten(), vehicles):
    sns.scatterplot(data=df, x="Traffic Situation", y=vehicle, ax=ax, palette="Set2")
    ax.set_title(f"{vehicle} vs. Traffic Situation")
plt.tight_layout()
plt.show()
```

# 8. Total Vehicle Count by Traffic Situation


```python
plt.figure(figsize=(10, 5))
sns.barplot(data=df, x="Traffic Situation", y="Total", estimator=sum, palette="coolwarm")
plt.title("Total Vehicle Count by Traffic Situation")
plt.show()
```

```python
df['Time'] = pd.to_datetime(df['Time'], errors='coerce')

df['Hour'] = df['Time'].dt.hour

plt.figure(figsize=(12, 6))
sns.barplot(data=df, x="Hour", y="Total", estimator=sum, palette="viridis")
plt.title("Busiest Hours of the Day for Traffic")
plt.xlabel("Hour of the Day")
plt.ylabel("Total Vehicle Count")
plt.show()
```
